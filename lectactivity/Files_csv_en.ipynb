### 3. Create new CSV files containing data only in the same year. Example: 2018.csv, 2017.csv, etc.

````python
import csv
unique_years = set()
 
with open ('annual-enterprise-survey-2018-financial-year-provisional-csv.csv', mode='r') as file:
    data=csv.DictReader(file)
    for row in data:
        year=row['Year']
        unique_years.add(year)
print(unique_years)
````

### Expected result:
```
{'2013', '2018', '2017', '2014', '2016', '2015'}
```

````python
for year in unique_years:
    with open(f"{year}.csv".format(year), "w") as ofile:
        # Начало вашего кода 
        writer = csv.writer(ofile)
        for row in data:
            if row[0]== year:
                writer.writerow(row)
        # Конец вашего кода
    print("{year}.csv is written".format(year),end=" ")
````
